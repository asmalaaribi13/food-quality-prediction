```python
def extract_intervalles(df):
    intervalles = {}
    for _, row in df.iterrows():
        produit = row['Produits']
        if produit not in intervalles:
            intervalles[produit] = {}
        

        protein_val = row['Protein_A_Ac']
        if isinstance(protein_val, str):
            max_val = float(protein_val[-2:])
            min_val = float(protein_val.split('<')[0])
            intervalles[produit]['Protein_A_Ac'] = (min_val, max_val)
        
      
        fine_val = row['Fine_NC']
        if isinstance(fine_val, str):
            max_val = float(fine_val)
            intervalles[produit]['Fine_NC'] = (0, max_val)
        
     
        durabilite_val = row['Durabilite_CA']
        if isinstance(durabilite_val, str):
            max_val = float(durabilite_val[-2:])
            min_val = float(durabilite_val.split('<')[0])
            intervalles[produit]['Durabilite_CA'] = (min_val, max_val)
        

        durete_val = row['Durete_CA']
        if isinstance(durete_val, str):
            max_val = float(durete_val[-2:])
            min_val = float(durete_val.split('<')[0])
            intervalles[produit]['Durete_CA'] = (min_val, max_val)
        

        amidon_val = row['Amidon_CA']
        if isinstance(amidon_val, str):
            max_val = float(amidon_val[-2:])
            min_val = float(amidon_val.split('<')[0])
            intervalles[produit]['Amidon_CA'] =  (min_val, max_val)
       
    
    return intervalles

intervalles = extract_intervalles(dCritere)

def detect_probleme(row, intervalles):
    produit = row['Type_de_produit']
    if produit in intervalles:
        problemes = []
        for col, (min_val, max_val) in intervalles[produit].items():
            if col == 'Fine_NC' and row['Fine'] > max_val:
                problemes.append('Taux de fine élevé')
            elif col == 'Durabilite_CA' and (row['Durabilite'] > min_val and row['Durabilite'] < max_val):
                problemes.append('Durabilité inférieur au cible')
            elif col == 'Durete_CA' and (row['Durete'] < min_val or row['Durete'] > max_val):
                problemes.append('Dureté inférieur au cible')
            elif col == 'Amidon_CA' and row['Amidon'] > max_val:
                problemes.append('Amidon supérieur au cible')
            elif col == 'Protein_A_Ac' and (row['Proteine'] > min_val and row['Proteine'] < max_val):
                problemes.append('Protéine non conforme')
            elif col == 'Amidon_CA' and row['Amidon'] < min_val:
                problemes.append('Amidon inférieur au cible')
        if problemes:
            return ', '.join(problemes)
        else:
            return 'Sans Probleme'
    else:
        return 'Sans Probleme'


df['Probleme'] = df.apply(lambda row: detect_probleme(row, intervalles), axis=1)
df['Probleme'] = df['Probleme'].apply(lambda x: x.replace('Sans Probleme, ', '') if x and 'Sans Probleme' in x else x)

df['Probleme']
```

The purpose of this code is to automate the process of identifying potential issues or non-conformities in product attributes based on predefined intervals. By flagging such problems, it helps in quality control and ensures that products meet specified standards.

- extract_intervalles function:
Iterates through each row of the DataFrame to extract intervals for specific attributes like protein content, fine rate, durability, hardness, and starch content for each type of product.
Stores these intervals in a dictionary named intervalles where the keys are product types, and the values are dictionaries containing attribute-interval pairs.
- detect_probleme function:
Takes a row of the DataFrame and the intervalles dictionary as input.
Checks if the product type exists in the intervalles dictionary.
Compares the actual values of attributes in the row with the corresponding intervals.
If any attribute falls outside the specified interval, it flags a potential problem associated with that attribute.
Returns a string describing any detected problems or 'Sans Probleme' if no problems are found.
- Applying detect_probleme function:
Uses the apply function to apply the detect_probleme function to each row of the DataFrame.
Adds a new column named 'Probleme' to the DataFrame containing the detected problems for each row.
